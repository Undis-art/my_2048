```python
from random import randint
from numpy import zeros

def occupied(numbers):
    occupied = []
    for n in numbers:
        occupied.append((n.row, n.col))
    return occupied

def empty(numbers):
    empty = []
    occupieds = occupied(numbers)
    if not occupieds:
        for i in range(4):
            for j in range(4):
                empty.append((i, j))
    else:
        for i in range(4):
            for j in range(4):
                if (i, j) not in occupieds:
                    empty.append((i, j))
    return empty

class Number:
    def __init__(self, value, row, col):
        self.value = value
        self.row = row
        self.col = col
        self.merged = False
    
    def at_edge(self, side, up):
        if side == -1 and self.col == 0:
            return True
        elif side == 1 and self.col == 3:
            return True
        elif up == 1 and self.row == 3:
            return True
        elif up == -1 and self.row == 0:
            return True
        else:
            return False
    
    def move(self, direction):
        side, up = direction
        while(not self.at_edge(side, up)):
            if (self.row + up, self.col + side) in occupied(numbers):
                if not self.merge(side, up):
                    break
            else:
                self.row = self.row + up
                self.col = self.col + side

    
    def merge(self, side, up):
        for n in numbers:
            if (n.row == self.row + up and 
                n.col == self.col + side and
                n.value == self.value and
                n.merged == False):
                n.value = n.value * 2
                numbers.pop(numbers.index(self))
                n.merged = True
                return True
        return False

numbers = []
    
def tupledir(key):
    if key == "a":
        direction = (-1, 0)
    elif key == "d":
        direction = (1, 0)
    elif key == "w":
        direction = (0, -1)
    elif key == "s":
        direction = (0, 1)

    return(direction)


def printmat():
    mat = zeros((4, 4), dtype=int)
    for n in numbers:
        mat[n.row][n.col] = n.value
    for n in range(4):
        for m in range(4):
            print(f'{mat[n][m]:3}', end = "")
        print("")
    print()
    
def addnumber():
    zero_places = empty(numbers)
    new_place = zero_places[randint(0, len(zero_places) - 1)]
    numbers.append(Number(randint(1, 2), new_place[0], new_place[1]))

def move(key):
    if key == "a":
        numbers.sort(key=lambda x: x.col)
        sorted_numbers = numbers.copy()
        for n in sorted_numbers:
            n.move(tupledir(key))
    if key == "d":
        numbers.sort(key=lambda x: x.col, reverse=True)
        sorted_numbers = numbers.copy()
        for n in sorted_numbers:
            n.move(tupledir(key))
    if key == "w":
        numbers.sort(key=lambda x: x.row)
        sorted_numbers = numbers.copy()
        for n in sorted_numbers:
            n.move(tupledir(key))
    if key == "s":
        numbers.sort(key=lambda x: x.row, reverse=True)
        sorted_numbers = numbers.copy()
        for n in sorted_numbers:
            n.move(tupledir(key))
    for n in numbers:
        n.merged = False
        
def play():
    mat = zeros((4, 4), dtype=int)
    numbers.clear()
    while True:
        addnumber()
        printmat()
        while True:
            key = input("use wasd. q quits.")
            if key in ("w", "a", "s", "d", "q"):
                break
        if key == "q":
            break
        else:
            move(key)

play()
```
